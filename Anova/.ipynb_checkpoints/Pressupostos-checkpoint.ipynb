```r
# Aplicacao cartesiana de vetores de nomes e dados sobre a funcao func, retorna um data frame "matricial" linear
crossDataFrame <- function(arra, arrb, data_a, data_b, func)
{
    x = 1:(length(arra)*length(arrb))
    y = 1:(length(arra)*length(arrb))
    z = 1:(length(arra)*length(arrb))
    
    for(i in 1:length(arra))
    {
        for(j in 1:length(arrb))
        {
            x[(i-1)*length(arra) + j] = arra[i]
            y[(i-1)*length(arra) + j] = arrb[j]
            z[(i-1)*length(arra) + j] = func(data_a[i], data_b[j])
        }
    }
    
    data.frame(x=x, y=y, z=z)
}

# Aplicacao linear de uma funcao sobre um vetor, retorna um data frame linear
linearDataFrame <- function(namea, names_b, arr, func)
{
    x = 1:(length(names_b))
    z = 1:(length(names_b))
    
    for(i in 1:length(arr))
    {
        x[i] = namea
        z[i] = func(arr[i])
    }
    
    data.frame(x=x, y=names_b, z=z)
}

# Transforma um dataframe em um vetor de colunas, sim
dfArrayCols <- function(df)
{
    x = vector(mode = "list", length = ncol(df)) 
    for(i in 1:ncol(df))
    {
        x[i] = list(df[,i])
    }
    x
}

# gera mapa de calor sobre um dataframe (x, y, z)
heatMap <- function(map)
{
    heat <- ggplot(map, aes(x, y)) + geom_tile(aes(fill = z))
    heat = heat + scale_fill_gradient(low = "#0000FF", high = "#EE6600") + theme(axis.title.y=element_blank(), axis.title.x=element_blank(), axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, size=16, face="bold"), axis.text.y = element_text(vjust = 0.5, hjust=1, size=16, face="bold"))
    heat
}

## Debug
#summary(lm(base$Preço_do_Produto ~ base$Unidades_Vendidas))$r.squared

func <- function(a, b)
{
    min(a,b) + max(a,b) + mean(a,b)
}
m = crossDataFrame(1:ncol(base), 1:ncol(base), 1:ncol(base), 1:ncol(base), func)
heatMap(m)

func <- function(a, b)
{
    options(warn=-1)
    a = unlist(a)
    b = unlist(b)
    fit <- summary(lm(a ~ b))$r.squared
    options(warn=0)
    fit
}
df <- dfArrayCols(base)
#m = crossDataFrame(colnames(base), colnames(base), df, df, func)
m = crossDataFrame(1:ncol(base), 1:ncol(base), df, df, func)
heatMap(m)

func <- function(a)
{
   a = unlist(a)
   shapiro.test(a)$statistic
}
m = linearDataFrame("Shapiro-Wilk", 1:ncol(base), df, func)
## 
```
