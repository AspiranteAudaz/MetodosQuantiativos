```r
#install.packages("tidyverse") 
library(tidyverse)
setwd("./");
```

```r
base = read.csv("./summer-products-with-rating-and-performance_2020-08.csv", header = TRUE)
```

```r
base_filtrada = select(base, price, units_sold, rating, rating_count, retail_price, merchant_rating, merchant_rating_count)
```

```r
colnames(base_filtrada)[0] <- "Produto"
colnames(base_filtrada)[1] <- "Preço_do_Produto"
colnames(base_filtrada)[2] <- "Unidades_Vendidas"
colnames(base_filtrada)[3] <- "Média_da_Nota_do_Produto"
colnames(base_filtrada)[4] <- "Quantidade_de_Avaliações"
colnames(base_filtrada)[5] <- "Preço_de_Varejo"
colnames(base_filtrada)[6] <- "Média_de_Nota_do_Vendedor"
colnames(base_filtrada)[7] <- "Quantidade_de_Avaliações_do_Vendedor"
```
