```r
library("dplyr")
library("ggplot2")  
library("moments")
```

```r
dataset <- read.csv(file="summer-products-with-rating-and-performance_2020-08.csv", header=TRUE, sep=",")
dataset <- as.data.frame(dataset)
base    <- select(dataset, price, units_sold, rating, rating_count, retail_price,  merchant_rating, merchant_rating_count)

# Nomes alternativos
columns = c("Price", "Units Sold", "Rating", "Rating C", "Retail Price", "Merch R", "Merch C")
colnames(base) <- columns
```

```r
inverse = function(x) {
    return (1/x)
}

for (c in columns) {    
    #print("----------------------------------------------")
    
    p = base[[c]]
    #plot(density(p))
    #print(p)
    #s = skewness(p, na.rm = TRUE)
    #print(s)
    
    print(paste("Shapiro test for ", c))
    a = shapiro.test(p)
    print(a)
    
    p2 = `^`(p, 2)
    print(paste("Shapiro test for ", c, "^2"))
    #print(p2)
    a = shapiro.test(p2)
    print(a)
    
    p_2 = sqrt(p)
    print(paste("Shapiro test for sqrt(", c, ")"))
    #print(p_2)
    a = shapiro.test(p_2)
    print(a)
    
    plog10 = log10(p)
    print(paste("Shapiro test for log10(", c, ")"))
    #print(plog10)
    a = shapiro.test(plog10)
    print(a)
    
    pinv = sapply(p, inverse)
    print(paste("Shapiro test for inv(", c, ")"))
    #print(pinv)
    a = shapiro.test(pinv)
    print(a)
}
```
